```python
import requests
from bs4 import BeautifulSoup
import urllib.request as urllib2
import re
import csv
```

```python
URL = "http://www.footballnews.com.au/forum/search.php?st=0&sk=t&sd=d&sr=posts&keywords=Ronaldo"
# page = requests.get(URL)
# print "Reading page..."
page = urllib2.urlopen(URL)
soupRonaldo = BeautifulSoup(page)
```

```python
url_base = "http://www.footballnews.com.au/forum/search.php?st=0&sk=t&sd=d&sr=posts&keywords=Ronaldo"
url_add = "&start="
loopCount = 0
numPosts = 30
textList_final = []
dateList_final = []
repliesList_final = []
viewsList_final = []
```

```python
detectTextCounter = 30
while detectTextCounter == 30:
    if loopCount == 0:
        url_final = url_base
    else:
        url_final = url_base+url_add+str(numPosts)
        numPosts += 30
        
    page = urllib2.urlopen(url_final)
    soupRonaldo = BeautifulSoup(page)
    divcontent_raw = soupRonaldo.find_all("div", class_="content")
    dateList = soupRonaldo.find_all("dd", class_="search-result-date")
    ddList = soupRonaldo.find_all("dd")
    detectTextCounter = len(divcontent_raw)
    
    nonNoneCount = 0
    while nonNoneCount == 0:
        for div in divcontent_raw:
            if div.blockquote != None:
                nonNoneCount += 1
                div.blockquote.decompose()
            if div.span != None:
                nonNoneCount += 1
    #             print(nonNoneCount)
                div.span.unwrap()
            if div.img != None:
                nonNoneCount += 1
                div.img.decompose()
        for dd in ddList:
            if dd.strong != None:
                nonNoneCount += 1
                dd.strong.unwrap()
            if dd.a != None:
                nonNoneCount += 1
                dd.a.decompose()
        if nonNoneCount != 0:
            nonNoneCount = 0
        else:
            break
            
    repliesList_raw = []
    viewsList_raw = []
    for i in range(len(ddList)):
        if i == 0:
            continue
        if "Replies" in ddList[i].get_text():
            repliesList_raw.append(ddList[i].get_text())
        if "Views" in ddList[i].get_text():
            viewsList_raw.append(ddList[i].get_text())
            
    textList_raw = []
    for div in divcontent_raw:
        textList_raw.append(div.get_text().replace("\n", "").replace("    ", " "))
        
#     textList_url= []
    for i in range(len(textList_raw)):
#         if i != 0:
#             if textList[i] != textList[i-1]:
#     #             print(textList[i])
#                 textlist_url.append(textList[i])
        
#     for i in range(len(textlist_url)):
        result = re.sub(r'http\S+', '', textList_raw[i], flags=re.MULTILINE)
        textList_final.append(result)
        
    for i in range(len(repliesList_raw)):
        repliesList_final.append(int(repliesList_raw[i].split(": ")[1]))
        
    for i in range(len(viewsList_raw)):
        viewsList_final.append(int(viewsList_raw[i].split(": ")[1]))
    
    for date in dateList:
        dateList_final.append(date.get_text())
    
    print(url_final)
    loopCount += 1
```

```python
textList_final_noDup = []
dateList_final_noDup = []
repliesList_final_nodup = []
viewsList_final_noDup = []
for i in range(len(textList_final)):
    if textList_final[i] not in textList_final_noDup:
        textList_final_noDup.append(textList_final[i])
        dateList_final_noDup.append(dateList_final[i])
        repliesList_final_nodup.append(repliesList_final[i])
        viewsList_final_noDup.append(viewsList_final[i])
```
