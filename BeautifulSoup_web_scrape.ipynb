Importing libraries

```python
import requests
from bs4 import BeautifulSoup
import urllib.request as urllib2
import re
import csv
```

Initializing variables

```python
# URL
url_ronaldo = "http://www.footballnews.com.au/forum/search.php?st=0&sk=t&sd=d&sr=posts&keywords=Ronaldo" # URL base for Cristiano Ronaldo
url_messi = "http://www.footballnews.com.au/forum/search.php?st=0&sk=t&sd=d&sr=posts&keywords=Messi" # URL base for Lionel Messi
url_add = "&start=" # URL additional for more than 30 posts

# for loop parts
loopCount_ronaldo = 0 # Loop counter for scanning Ronaldo-related posts
loopCount_messi = 0 # Loop counter for scanning Messi-related posts
numPosts_ronaldo = 30 # Number of Ronaldo-related posts
numPosts_messi = 30 # Number of Messi-related posts

# Lists
textList_final_ronaldo = [] # List of Ronaldo-related posts' texts
textList_final_messi = [] # List of Messi-related posts' texts
dateList_final_ronaldo = [] # List of Ronaldo-related posts' datetime
dateList_final_messi = [] # List of Messi-related posts' datetime
repliesList_final_ronaldo = [] # List of Ronaldo-related posts' replies number
repliesList_final_messi = [] # List of Messi-related posts' replies number
viewsList_final_ronaldo = [] # List of Ronaldo-related posts' views number
viewsList_final_messi = [] # List of Messi-related posts' views numberxts
```

```python
detectTextCounter = 30 # detecting how many posts in one page
while detectTextCounter == 30:
    if loopCount_ronaldo == 0: # if it's first page,
        url_final = url_ronaldo # Then only use the URL base
    else: # else
        url_final = url_ronaldo+url_add+str(numPosts_ronaldo) # Add the additional URL for more than 30 posts
        numPosts_ronaldo += 30 # Add more 30 in number of posts
        
    page = urllib2.urlopen(url_final) # Opening the page
    soupRonaldo = BeautifulSoup(page) # Web scraping the page
    divcontent_raw = soupRonaldo.find_all("div", class_="content") # Find all <div> tags that contains posts' texts
    dateList = soupRonaldo.find_all("dd", class_="search-result-date") # find all <dd> tags that contains posts' datetime
    ddList = soupRonaldo.find_all("dd") # find all <dd> tags
    detectTextCounter = len(divcontent_raw) # detecting how many posts in current page
    
    nonNoneCount = 0 # Counter for non-None results
    while nonNoneCount == 0:
        for div in divcontent_raw:
            if div.blockquote != None: # if <blockquote> tags in <div> tags exists
                nonNoneCount += 1
                div.blockquote.decompose() # Remove the <blockquote> tags along with the component in it
            if div.span != None: # if <span> tags in <div> tags exists
                nonNoneCount += 1
                div.span.unwrap() # Remove the <span> tags
            if div.img != None: # if <img> tags in <div> tags exists
                nonNoneCount += 1
                div.img.decompose() # Remove the <img> tags along with the component in it
        for dd in ddList:
            if dd.strong != None: # if <strong> tags in <dd> tags exists
                nonNoneCount += 1
                dd.strong.unwrap() # Remove the <strong> tags
            if dd.a != None: # if <a> tags in <d> tags exists
                nonNoneCount += 1
                dd.a.decompose() # Remove the <a> tags along with the component in it
        if nonNoneCount != 0:
            nonNoneCount = 0
        else:
            break
            
    repliesList_raw = [] # list of Ronaldo-related posts' replies number for current page
    viewsList_raw = [] # list of Ronaldo-related posts' views number for current page
    for i in range(len(ddList)):
        if i == 0:
            continue
        if "Replies" in ddList[i].get_text(): # if 'Replies' word is part of the text
            repliesList_raw.append(ddList[i].get_text()) # add the replies number to the list for current page
        if "Views" in ddList[i].get_text(): # if 'Views' word is part of the text
            viewsList_raw.append(ddList[i].get_text()) # add the views number to the list for current page
            
    textList_raw = []
    for div in divcontent_raw:
        # Removing every unnecessary parts of the posts' texts (such as enter, long whitespaces, 
        # leading and tailing whitespaces and '...')
        textList_raw.append(div.get_text().replace("\n", "").replace("    ", " ").replace("...", "").replace("  ", " ").replace("   ", " ").strip())
        
    for i in range(len(textList_raw)):
        result = re.sub(r'http\S+', '', textList_raw[i], flags=re.MULTILINE) # Remove every URL in the posts' texts
        textList_final_ronaldo.append(result) # Append the posts' texts in the final list
        
    for i in range(len(repliesList_raw)):
        # Append the posts' replies number in the final list
        repliesList_final_ronaldo.append(int(repliesList_raw[i].split(": ")[1]))
        
    for i in range(len(viewsList_raw)):
        # Append the posts' views number in the final list
        viewsList_final_ronaldo.append(int(viewsList_raw[i].split(": ")[1]))
    
    for date in dateList:
        # Append the posts' datetime in the final list
        dateList_final_ronaldo.append(date.get_text())
    
    print(url_final) # print the URL destination
    loopCount_ronaldo += 1 # adding counter for number of page scraped
```

```python
textList_final_ronaldo_noDup = [] # List of Ronaldo-related posts' text with no duplicates
dateList_final_ronaldo_noDup = [] # List of Ronaldo-related posts' datetime with no duplicates
repliesList_final_ronaldo_nodup = [] # List of Ronaldo-related posts' replies number with no duplicates
viewsList_final_ronaldo_noDup = [] # List of Ronaldo-related posts' views number with no duplicates
for i in range(len(textList_final_ronaldo)):
    if textList_final_ronaldo[i] not in textList_final_ronaldo_noDup: # if text does not exist in the no-duplicate list
        # Added to the posts' texts' no-duplicate list
        textList_final_ronaldo_noDup.append(textList_final_ronaldo[i])
        # Added to the posts' datetime' no-duplicate list
        dateList_final_ronaldo_noDup.append(dateList_final_ronaldo[i])
        # Added to the posts' replies numbers' no-duplicate list
        repliesList_final_ronaldo_nodup.append(repliesList_final_ronaldo[i])
        # Added to the posts' views numbers' no-duplicate list
        viewsList_final_ronaldo_noDup.append(viewsList_final_ronaldo[i])
```
